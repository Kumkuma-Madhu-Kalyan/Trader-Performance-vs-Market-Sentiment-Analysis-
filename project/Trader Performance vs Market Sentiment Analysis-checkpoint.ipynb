# sentiment dataset terminology

- Timestamp >> Time of sentiment (Used to match with trades)
- classification >> market sentiment or investor sentiment (categories)
    - Fear = Market is scared (People sell or avoid buying)
    - Extreme Fear = Panic situation (Heavy selling, prices often low)
    - Greed = Market is confident (People start buying)

    - Neutral = Balanced sentiment or Market undecided (trade carefully or wait)
- Date >> Human-readable date
- Value (0–100) >> Fear/Greed score

# trades dataset terminology

**Basic Trade Details**
- Account >> The digital wallet address (ID) that made the trade.
- Coin >> Asset traded (Which crypto/token or various  cryptocurrencies)
- Execution Price >> The actual price per coin at which the trade was completed.
- Size Tokens >> The exact number of coins you bought or sold in that specific transaction.
- Size USD >> The total value of those tokens in US Dollars at the time of the trade.
    - Calculation: Execution Price × Size Tokens = Size USD

**Trade Action**
- Side >> Whether you are BUYing or SELLing.
    - BUY = expecting price to go UP
    - SELL = expecting price to go DOWN
- Timestamp IST >> Time of trade (Match with sentiment)
- Direction >> Trade intention(Strategy direction)
    - Buy = Purchasing an asset (like stock or crypto) to own it, expecting the price to rise.
    - Sell = Getting rid of an asset you own, or initiating a short position to profit from a price drop.
    - Open Long = Buying an asset with the intention of holding it because you believe the price will increase (being bullish).
    - Close Long = Selling the asset you previously bought to lock in profits or losses.
    - Spot Dust Conversion = The process of converting tiny, unusable leftover amounts of crypto (dust) into a platform’s native token.
    - Open Short = Selling borrowed assets, hoping to buy them back later at a lower price to make a profit.
    - Close Short = Buying back the borrowed assets to settle the short position.
    - Short > Long = Market sentiment is bearish; more people are betting on prices to fall than rise.
    - Auto-Deleveraging (ADL) = A risk management system in futures trading that automatically closes profitable positions if a liquidated account's loss is too massive to be covered by the insurance fund.
    - Liquidated Isolated Short = An isolated short position is forcibly closed by the exchange because the price rose too high, and the loss exceeded the specific collateral (margin) set aside for that single trade.
    - Settlement = The final process where the exchange settles up, calculating profits and losses and transferring them, often at a specific time in futures

**Position & Strategy**
- Start Position >> How many tokens you already held before this specific line item was processed. You can see this number grow as more "BUY" orders are filled.
- Crossed >> This usually indicates a "Market" order that "crossed" the spread to fill immediately against an existing seller. TRUE means you took liquidity from the market.
- Order ID >> Order number (Track orders)

**Profit & Cost**
- Closed PnL >> Profit or Loss (Performance)
- Fee >> The amount paid to the exchange or network to process the trade

    - PnL > 0 → Profit ✅
    - PnL < 0 → Loss ❌

**Technical**
- Transaction Hash >> The unique hash/ID of the transaction on the blockchain.
- Trade ID >> A unique ID for every individual small "fill" that makes up the larger order.
- Timestamp (scientific) >> The exact date and time the trade happened.

## Q) Does performance (PnL, win rate, drawdown proxy) differ between Fear vs Greed days?

1) Performance vs Sentiment

```python
performance = merged.groupby('classification')['Closed PnL'].agg(['mean','sum','count'])
performance
```

```python
merged.groupby('classification')['Closed PnL'].mean().plot(kind='bar')
plt.title("Average PnL by Sentiment")
plt.show()
```

**Key Observations**

- Peak Performance: The highest average Pnl performance occurs Extreme Greed sentiment.
- when compared to Fear and Greed sentiment , we can observe Average Pnl performace is double the Greed sentiment, that indicated traders are not taking any risk, they are earning their profits by selling the tokens when the market is scared (fear). 

2) Win rate vs Sentiment

```python
win_rate = merged.groupby('classification')['win'].mean()
print(win_rate)
```

```python
# Chart
win_rate.plot(kind='bar')
plt.title("Win Rate accorss the sentiments")
plt.ylabel("Win Rate")
plt.show()
```

**key observation**
- similar to above here also the highest win ratio is observed in Extreme Greed sentiment.
- when compared to Fear and Greed sentiments, trades and making more profits in Greed sentiment by buying more Tokens when the market is confident and over confident
- as sentiment shifts from Fear toward Greed, the probability of a "Win" generally increases, with the most favorable conditions occurring when the market is in a state of Extreme Greed. 

3) Drawdown accross the sentiments

drawdown which measures the capital loss or risk of a strategy over a specific period

```python
loss_data = merged[merged['Closed PnL'] < 0]

drawdown = loss_data.groupby('classification')['Closed PnL'].agg(['mean','min'])
drawdown
```

```python
# Chart (worst loss)
drawdown['min'].plot(kind='bar')
plt.title("Worst Loss (Drawdown Proxy)")
plt.ylabel("PnL")
plt.show()
```

**Key Observations**
- when the market situation is confident (Greed) , then most of the traders are buying the token that leads to heavy loss of 115000 USD .

**Conclusion**

- Trading performance differs significantly between Fear and Greed sentiments.
- During Fear, traders tend to achieve higher average PnL and better win rates, suggesting that lower prices provide better entry opportunities.
- During Greed, although trading activity increases, performance often declines and risk (drawdowns) increases due to overconfidence and larger position sizes.This indicates that market sentiment strongly impacts both profitability and risk-taking behavior.

## Q) Trader Segmentation

Segment 1: High vs Low Leverage

```python
# Average leverage per trader
lev_trader = merged.groupby('Account')['leverage_proxy'].mean().reset_index()
lev_trader.head()
```

```python
# Threshold (median split)
threshold = lev_trader['leverage_proxy'].median()
threshold
```

```python
lev_trader['leverage_group'] = lev_trader['leverage_proxy'].apply(
    lambda x: 'High Leverage' if x > threshold else 'Low Leverage'
)

# Merge back
merged = merged.merge(lev_trader[['Account','leverage_group']], on='Account')
```

```python
# Grouping
data = merged.groupby('leverage_group')['Closed PnL'].mean()

# Pie chart
plt.figure()
data.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title("PnL Distribution: High vs Low Leverage")
plt.ylabel('')  # removes the default ylabel
plt.show()
```

Segment 2: Frequent vs Infrequent Traders

```python
trade_counts = merged['Account'].value_counts()
trade_counts

threshold = trade_counts.median()
# threshold
merged['freq_group'] = merged['Account'].apply(
    lambda x: 'Frequent' if trade_counts[x] > threshold else 'Infrequent'
)

# merged['freq_group'] = merged['Account'].map(lambda x: 'High' if trade_counts[x]>10 else 'Low')
merged
```

```python
data = merged.groupby('freq_group')['Closed PnL'].mean()

# Pie chart
plt.figure()
data.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title("PnL Distribution: Frequent vs Infrequent Traders")
plt.ylabel('')
plt.show()
```

Segment 3: Consistent vs Inconsistent

```python
consistency = merged.groupby('Account')['Closed PnL'].std().reset_index()
threshold = consistency['Closed PnL'].median()

consistency['consistency_group'] = consistency['Closed PnL'].apply(
    lambda x: 'Consistent' if x < threshold else 'Inconsistent'
)

# Merge
merged = merged.merge(consistency[['Account','consistency_group']], on='Account')

# merged = merged.merge(consistency, on='Account', suffixes=('','_std'))
# merged['consistency_group'] = merged['Closed PnL_std'].apply(lambda x: 'Consistent' if x<50 else 'Volatile')
merged
```

```python
merged.groupby('consistency_group')['Closed PnL'].mean().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("PnL: Consistent vs Inconsistent Traders")
plt.show()
```

**Note**
- here standard deviation indicated How much a trader’s profits fluctuate
| Std Value | Meaning                                   |
| --------- | ----------------------------------------- |
| Low std   | Stable profits → **Consistent trader**    |
| High std  | Big ups & downs → **Inconsistent trader** |

- here be like a Proactive traders , During market is scared most of the traders will be in 'fear of crash' which inturn leads in selling the trades this will add advantage of buys trades in low cost and make profits.

**STRATEGY 2:"Stay calm and play it safe, even when everyone else is rushing in"**

When the market is booming and everyone is making money, traders often get "greedy." They start taking bigger risks, like betting too much money on a single trade or borrowing extra cash (leverage) to try and win big.

The problem is that this usually backfires. Because they are taking such big risks, they end up losing more money than they expected when the market eventually dips.

To stay safe, it’s better to use a defensive strategy:
- Bet smaller: Don't put all your eggs in one basket.
- Trade less: Don't feel the need to jump on every single trend.
- Protect your cash: Focus on keeping what you have so you’re ready when the market gets tough.
